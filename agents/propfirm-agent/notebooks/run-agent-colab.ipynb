# Prop Firm Agent - Free Colab Runner
تشغيل مجاني بديل عن GitHub Actions. شغّل الخلايا بالترتيب من 1 إلى 6.

```python
import os, subprocess
os.chdir('/content')
subprocess.run('rm -rf Compari-prop-firm && git clone https://github.com/l7adj/Compari-prop-firm.git', shell=True, check=True)
os.chdir('/content/Compari-prop-firm')
print('PROJECT_DIR=', os.getcwd())
```

```python
import subprocess
subprocess.run('npm install', shell=True, check=True)
subprocess.run('npx playwright install --with-deps chromium', shell=True, check=True)
print('Dependencies installed')
```

```python
import subprocess
subprocess.run('npm run agent:schema -- --bundle agents/propfirm-agent/schema/schema-bundle --outSchema agents/propfirm-agent/schema/schema-driver.generated.json --outCompanySeed agents/propfirm-agent/companies/company-seed.generated.json --outPropSeed agents/propfirm-agent/companies/prop-firms-seed.generated.json', shell=True, check=True)
print('Schema driver generated')
```

```python
import subprocess
subprocess.run('rm -rf output && npm run agent:batch -- --batch agents/propfirm-agent/companies/prop-firms-seed.generated.json --limit 5 --offset 0 --categories cfd_prop_firm,futures_prop_firm,crypto_prop_firm', shell=True, check=True)
print('Batch scrape finished')
```

```python
import subprocess
subprocess.run('npm run agent:review -- --input output/normalized_pending_review.csv --output output --schema agents/propfirm-agent/schema/schema-driver.generated.json', shell=True, check=True)
print('Schema review finished')
```

```python
import os, subprocess
from google.colab import files
subprocess.run('zip -r propfirm-agent-output.zip output agents/propfirm-agent/schema/schema-driver.generated.json agents/propfirm-agent/companies/prop-firms-seed.generated.json', shell=True, check=True)
print('ZIP ready:', os.path.abspath('propfirm-agent-output.zip'))
files.download('propfirm-agent-output.zip')
```
